Tipo de Teste de Hipótese a ser feito: Unilateral
- Conheço apenas os desvios amostrais dados.
- São amostras de populações diferentes: Pop 1: Antes da melhoria. Pop 2: Depois da melhoria.
- Quero trabalhar estes dados de forma a consiguir elaborar um teste de hipótese unilateral, haja vista que o objetivo é verificar se a média dos dados dos de tratamento é menor que a dos dados de controle.

Para isso, assumi-se que a média dos dados históricos é a média populacional, mas o desvio populacional é desconhecido.


Explicação para assumir que a média encontrada para o controle é a média populacional. Fazendo o intervalo de confiança utilizando o **modelo normal** e a **t de Student**.



```python
x=np.arange(0,10000,0.01)                                      # Decalrando um intevalor de valores de x para poder plotar gráficos.
u_pop= mean_controle                                           # Assumindo que a  média populacional = média do controle.
u_amostras=mean_tratamento                                     # Assumindo que os dados de tratamento são amostras para fazer o teste de hipótese.
sigma_amostra=sigma_tratamento                                 # Nomeando o desvio padrão de tratamento como o desvio padrão da amostra.

# Para validar a escolha dos dados de controle como uma população, faz-se um intervalo de confiança para a média populacional de controle. 

#Utilizando a t de Student

# Nível de Confiança:
nc=0.90

# Definindo o número de graus de liberdade, como sendo o tamanho da amostra - 1
grau_liberdade_controle=len(dados_controle)-1

# Valor do 't de Student' com base no nível de confiança e no grau de liberdade
t_controle= sc.t.ppf(1-(1-nc)/2,grau_liberdade_controle)

# Estabelecendo os limites superior e inferior do Intervalo de Confiança, seguindo a 't de Student'
limite_p,limite_m=u_pop+t_controle*sigma_controle/np.sqrt(len(dados_controle)),u_pop-t_controle*sigma_controle/np.sqrt(len(dados_controle))  
print('Nível de confiança',nc*100,"%\n")
print("Distribuição T de Student")
print("Função t: ",t_controle)
print("Média populacional",u_pop)
print(f'Limite do intervalo de confiança: [{limite_m},{limite_p}]')
print()


#utilizando normal
print("Distribuição Normal")

# Valor da Distribuição Normal Z com base no nível de confiança 
norm_controle= sc.norm.ppf(1-(1-nc)/2,0,1)

# Estabelecendo os limites superior e inferior do Intervalo de Confiança, seguindo a Distribuição Normal
limiten_p,limiten_m=u_pop+norm_controle*sigma_controle/np.sqrt(len(dados_controle)),u_pop-norm_controle*sigma_controle/np.sqrt(len(dados_controle))  
print("Função z: ",norm_controle)
print("Média Pop: ",u_pop)
print(f'Intercalo de confiança: [{limiten_m},{limiten_p}]')

print(f"\n Perce-se uma semelhança considerável entre os intervalos, assim como uma precisão grande para assumir que \n a média populacional é {round(u_pop,3)}.")

#Pelo número de amostras posso aproximar que os dados de controle correspondem aos dados populacionais, sobretudo por conta do intervalo de confiança assumindo um desvio populacional(Distribuição Normal) e amostral(Distribuição T-Student) serem semelhantes.
```

```python
#Outro teste para entender melhor qual teste é o melhor a ser utilizado: O comparação entre médias ou o teste unilateral.

print("Teste para validar se o problema envolve uma ou duas populações.\n")
valor_conjunto= (u_amostras-u_pop)/np.sqrt((sigma_amostra**2/len(dados_tratamento))+(sigma_controle**2/len(dados_controle))) #cálculo da estatística do teste para duas populações.
print("Considerando duas populações: z= ",valor_conjunto)
valor_separado=(u_amostras-u_pop)/np.sqrt((sigma_amostra**2/len(dados_tratamento))) #cálculo da estatística do teste para uma população.
print("Considerando população de tratamento apenas: z= ",valor_separado)

print("Diferença percentual: ",(1-(valor_conjunto/valor_separado))*100,"%") #cálculo da variação percentual entre as estatísticas.

print("\nPosso considerar para o problema apenas uma população (a de tratamento) para o cálculo da estatística do teste.")
print("Portanto, utiliza-se um Teste Unilateral no problema.")
```

Evidenciando as expressões, de maneira visual, dos Intervalos de Confiança  para **Distribuição Normal** e  **t de Student**, sendo



*   m = Média amostral
*  z (alpha/2) = Acumulada da função normal

*   sigma(x) = Desvio populacional
*   n = Número de amostras

*   t(alpha/2, n-1) = Acumulada da função 't de Student'
*   s = Desvio amostral

A seguir, realiza-se o Teste de Hipótese a fim de compreender-se se a hipótese nula H0 vai ser ou não rejeitada para cada nível de significância assinalado. Nesse caso, se o valor de 't' da estatística do teste for maior que a da significância, a hipótese nula não será rejeitada. Caso contrário, a hipótese nula poderá ser rejeitada em favor da nova hipótese alternativa HA.

Em qualquer um dos casos, deverá ser testado valores de níveis de significância, nomeados por 'a' no código a seguir, de maneira que com alguns níveis haverá a rejeição e em outros não.

```python
print("Hipótese nula: O novo tempo depois da melhoria não é menor que o antigo!\n")
print("Hipótese Alternativa: O novo tempo depois da melhoria é menor que o antigo!\n")
print(f"H0 => upop_tratamento = {u_pop}")
print(f'HA => upop_tratamento < {u_pop}\n')

t=(u_amostras-u_pop)/(sigma_tratamento/np.sqrt(len(dados_tratamento))) #estatística do teste seguindo uma distribuição t.

print(f'Função t (Estatística do Teste): {t}\n')

# Estabelecendo possíveis níveis de significância
a1= 0.01 # 1%
a2=0.05  # 5%
a3=0.1   # 10%

grau_liberdade=len(dados_tratamento)-1 #estabelecendo o grau de liverdade n-1.

t_est1=sc.t.ppf(a1,grau_liberdade)     #calculando a função t para cada significância e fazendo o este de hipótese para cada uma.
print(f'Alpha: {a1}, t ={t_est1}')
if t>t_est1:
  print('Não rejeito a hipótese nula.\n')
else:
  print("Rejeito a hipótese nula.\n")

t_est2=sc.t.ppf(a2,grau_liberdade)
print(f'Alpha: {a2}, t={t_est2}')
if t>t_est2:
  print('Não rejeito a hipótese nula.\n')
else:
  print("Rejeito a hipótese nula.\n")

t_est3=sc.t.ppf(a3,grau_liberdade)
print(f'Alpha: {a3}, t={t_est3}')
if t>t_est3:
  print('Não rejeito a hipótese nula.')
else:
  print("Rejeito a hipótese nula.")

#Graficando os resultados. 

x = np.arange(-5,5,0.5)
plt.figure(figsize=[15,7])
plt.title("Representação gráfica de cada significância em comparação à estatística do teste",fontsize="15")
plt.xticks(x)
plt.scatter(t,0,c="r",label="Estatística do teste"+"  ;t: "+str(round(t,3)))
plt.scatter(t_est1,0,c="b",label="Alpha: "+str(a1)+"  ;t: "+str(round(t_est1,3)))
plt.scatter(t_est2,0,c="y",label="Alpha: "+str(a2)+"  ;t: "+str(round(t_est2,3)))
plt.scatter(t_est3,0,c="m",label="Alpha: "+str(a3)+"  ;t: "+str(round(t_est3,3)))
plt.legend(fontsize=15)
     
x = np.arange(-5,5,0.01)  
     
y1 = sc.t.pdf(x,219,0)   
plt.axvline(0,0,1,color="b",ls="--")
plt.plot(x, y1, "-")  
plt.show()
```

Nesse momento busca-se a obtenção do 'valor-p', ou seja, o mínimo de significância a partir do qual a hipótese nula será rejeitada.

```python
#Criando uma função que grafica os testes de hipótese para cada significância e mostra visualmente a rejeição ou não da hipótese nula.

def plot(est,alpha):
  
  t= t=(u_amostras-u_pop)/(sigma_tratamento/np.sqrt(len(dados_tratamento)))
  x = np.arange(-5,5,0.5)
  plt.figure(figsize=[15,7])
  plt.title(f"Representação gráfica do Teste para significância de {alpha*100}%",fontsize=15)
  plt.xticks(x)
  plt.scatter(t,0,c="r",label="Teste:"+"  ;t: "+str(round(t,3)))
  plt.axvline(t,0,0.5,color="r",ls="--")
  plt.scatter(est,0,c="b",label="Alpha: "+str(alpha)+"  ;t: "+str(round(est,3)))
  plt.axvline(est,0,0.5,color="b",ls="--")

  plt.legend(fontsize=15)
      
  x = np.arange(-5,5,0.01)  

  x2=np.arange(-5,est,0.01)
  plt.bar(x2,sc.t.pdf(x2,219,0),width=0.01,color="black")

  x3=np.arange(est,5,0.01)
  plt.bar(x3,sc.t.pdf(x3,219,0),width=0.01,color="gray")

  y1 = sc.t.pdf(x,219,0)   
  plt.plot(x, y1, "-")  

  plt.arrow(est-0.3,0.02,-2,0,width=0.003,head_width=0.03,color="black")
  plt.text(-5,0.045,f"({round((alpha)*100,3)}%)",fontsize=12,backgroundcolor='0.9')
  plt.text(-5,0.07,f"Área de Rejeição de H0;",fontsize=12,backgroundcolor='0.9')
  plt.text(-0.3,0.02,f"({round((1-alpha),3)*100}%)",fontsize=12,backgroundcolor='0.9')
  plt.text(-1,0.1,f"Área de Não-Rejeição de H0;",fontsize=12,backgroundcolor='0.9')
 
  plt.legend(fontsize=15)
  return plt.show()
```

```python
plot(t_est1,a1) #chamando a função para a1.
```

```python
plot(t_est2,a2) #chamando a função para a2.
```

```python
plot(t_est3,a3) #chamando a função para a3.
```

```python
#Valor-p
pvalor = sc.t.cdf(t, df = grau_liberdade) #calculando o valor p do teste.
print("O Valor p do teste é: ",pvalor*100,"%\n")

#Graficando os resultados.

plt.figure(figsize=[15,7])
x = np.arange(-5,5,0.5)
plt.xticks(x)
x = np.arange(-5,5,0.01)
y1 = sc.t.pdf(x,219,0)   
plt.axvline(0,0,1,color="b",ls="--")
plt.plot(x, y1, "-")  
x2=np.arange(-5,t,0.01)

plt.scatter(t,0.001,c="r",label="Valor-P"+"  ;t: "+str(round(t,3)))
plt.bar(x2,sc.t.pdf(x2,219,0),width=0.01)
plt.title("Representação do valor-p graficamente",fontsize=15)

plt.legend(fontsize=15)
plt.text(-3,0.02,f"({round((pvalor)*100,3)}%)",fontsize=10,backgroundcolor='0.9')
plt.show()
```

Visualiza-se com o gráfico acima um 'valor-p', ou seja, o mínimo de significância a partir da qual a hipótese nula é rejeitada. Para este teste, o valor p é pequeno (6,203%).

 Esse valor significa que o teste de hipótese indica uma **rejeição da hipótese nula** em relação à nova proposta. Isso, porque um 'valor-p' pequeno mostra que a chance de cometer um erro de tipo-1 (rejeitar a hipótese nula sendo que essa é a verdadeira) é pequena, resultando, portanto, num favorecimento em relação à sua rejeição e troca pelo novo sistema criado.

 A saber, 'valores-p' mais expressivos, indicam que a probabilidade de rejeitar a hipótese nula, quando ela expressa a realidade, é maior, ou seja, não é algo "seguro" fazer a troca da hipótese nula pela nova, mantendo-a, portanto.

```python
#Graficando uma comparação entre o valor p e as significâncias.

x = np.arange(-5,5,0.5)
plt.figure(figsize=[15,7])
plt.title("Comparação do Valor-p com as significâncias",fontsize=15)
plt.xticks(x)
plt.axvline(t,0,0.5,color="r",ls="--")
plt.scatter(t,0,c="r",label="Valor-p: "+str(round(pvalor,3))+"  ;t: "+str(round(t,3)))

plt.scatter(t_est1,0,c="b",label="Alpha: "+str(a1)+"  ;t: "+str(round(t_est1,3)))
plt.axvline(t_est1,0,0.5,color="b",ls="--")

plt.scatter(t_est2,0,c="y",label="Alpha: "+str(a2)+"  ;t: "+str(round(t_est2,3)))
plt.axvline(t_est2,0,0.5,color="y",ls="--")

plt.scatter(t_est3,0,c="m",label="Alpha: "+str(a3)+"  ;t: "+str(round(t_est3,3)))
plt.axvline(t_est3,0,0.5,color="m",ls="--")

plt.legend(fontsize=15)
     
x = np.arange(-5,5,0.01)  

x2=np.arange(-5,t,0.01)
plt.bar(x2,sc.t.pdf(x2,219,0),width=0.01,color="black")

x3=np.arange(t,5,0.01)
plt.bar(x3,sc.t.pdf(x3,219,0),width=0.01,color="gray")

y1 = sc.t.pdf(x,219,0)   
plt.plot(x, y1, "-")  

plt.arrow(t-0.3,0.02,-2,0,width=0.003,head_width=0.03,color="black")
plt.text(-5,0.02,f"({round((pvalor)*100,3)}%)",fontsize=12,backgroundcolor='0.9')
plt.text(-5,0.07,f" Alpha<valor-p",fontsize=12,backgroundcolor='0.9')
plt.text(-0.3,0.02,f"({round((1-pvalor)*100,3)}%)",fontsize=12,backgroundcolor='0.9')
plt.text(-1,0.1,f"Alpha>valor-p",fontsize=12,backgroundcolor='0.9')
plt.show()
```

Concluindo o problema, evidencia-se que a alteração dos níveis de significância(a) entre 1%,5% e 10%, faz com que a hipótese nula seja rejeitada ou não. Assim, para 1% e 5% nota-se que não é vantajoso rejeitar a hipótese nula, já para 10% já é recomendado a troca. Isso está de acordo com o 'valor-p' encontrado anteriormente (6,203%), ou seja, o mínimo valor de significância para que a hipótese nula seja rejeitada está acima dos valores que não rejeitam (1% e 5%), e evidencia que o valor de 10% possibilita a alteração, já que 6,203% < 10% .

Essas informações também coincidem com aquilo que é visualizado nos gráficos finais (rejeição ou não da hipótese nula) a partir de níveis de significância distintos. De maneira que é possível visualizar a localização dos valores 'a' em relação ao 'valor-p', ou seja, se o 'a' encontra-se' à esquerda do 'valor-p' rejeita-se a hipótese nula', caso contrário, ela não é rejeitada.
